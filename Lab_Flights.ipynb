```python
import pandas as pd
```

```python
from sqlalchemy import create_engine,MetaData, Table, select
from sqlalchemy.sql import text
```

```python
engine = create_engine('postgresql+psycopg2://postgres:S_1pLa2k@127.0.0.1/Lab_Flights')
```

```python
con = engine.connect()
```

```python
# 1.Read both tables using only SQLAlchemy (no Pandas). Use a separate command for each.
metadata = MetaData()

# Load the tables
table1 = Table('flights', metadata, autoload_with=engine)
table2 = Table('bookings', metadata, autoload_with=engine)

# Create a connection
with engine.connect() as connection:
    # Select all rows from table 1
    result_table1 = connection.execute(table1.select()).fetchall()
    
    # Select all rows from table 2
    result_table2 = connection.execute(table2.select()).fetchall()

# Print the results
print("Table 1:")
print(result_table1)

print("\nTable 2:")
print(result_table2)
```

```python
# 2. Create a Pandas Dataframe for each table and print them.
flights = "Select * from flights"
flights_df = pd.read_sql(flights,con)
flights_df.head()
```

```python
bookings = "Select * from bookings"
booking_df = pd.read_sql(bookings,con)
booking_df.head()
```

```python
# 3.Use an SQL JOIN based on the flight id to load both tables into a single Pandas Dataframe.
join = "select * from flights join bookings on flights.id = bookings.id"
join_df = pd.read_sql(join,con)
join_df.head()
```
